# Q15 - Input: Circle calculations

Ask the user to input the radius of a circle (float). Compute and print:
- Circumference = `2 * 3.14159 * radius`
- Area = `3.14159 * radius ** 2`

```python
# Answer here
# Hint: use input(), then convert to int() or float() as needed.
radius = float(input("Enter the radius of the circle "))
pi = 3.15159
circumference = 2*pi *radius
area = pi*radius**2
print(circumference)
print(area)
```

# Q16 - Input: Convert total seconds to hours, minutes, seconds

Ask the user to input a total number of seconds (integer). Convert it to hours, minutes, and seconds and print as `H:M:S`.

```python
# Answer here
# Hint: use input(), then convert to int() or float() as needed.
seconds = int(input("Enter total number of seconds you have "))
mins = int(seconds/60)
hours = int(mins/60)
left_second = seconds%60
left_mins = mins%60
print(hours, "H", left_mins , "M" , left_second , "S")



```

# Q17 - Input: Simple interest calculation

```python
# Answer here
principal = int(input("Enter your principal amount"))
rate = float(input("Enter your rate of interest "))
time = int(input("Enter number of Years "))
SI = (principal*rate*time) / 100
print("Your simple interest is :", SI)
```
